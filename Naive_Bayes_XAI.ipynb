```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import shap, lime.lime_tabular, numpy as np
import matplotlib.pyplot as plt


start_time_nb = time.time()

# Train
nb = GaussianNB()
nb.fit(X_train, y_train)

training_time_nb = time.time() - start_time_nb
```

```python


# Step 6: Predictions and Evaluation Function
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# Step 6: Predictions and Evaluation Function
def evaluate_model(model, X_test, y_test, name, training_time):
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)

    # Choose macro for balanced importance across all classes
    f1 = f1_score(y_test, y_pred, average='macro')

    cm = confusion_matrix(y_test, y_pred)

    # FPR and FNR are only reliable in binary, so we check dimensions
    if cm.shape == (2, 2):
        fp = cm[0][1] / (cm[0][0] + cm[0][1])
        fn = cm[1][0] / (cm[1][0] + cm[1][1])
    else:
        fp = fn = "N/A (multiclass)"

    print(f"\n{name} Evaluation:")
    print(f"Accuracy: {acc:.4f}")
    print(f"F1 Score (macro): {f1:.4f}")
    print(f"False Positive Rate (FPR): {fp}")
    print(f"False Negative Rate (FNR): {fn}")
    print(f"Training Time: {training_time:.4f} seconds")


# Evaluate all models
evaluate_model(nb, X_test, y_test, "Naive Bayes", training_time_nb)
```

```python
# Feature Likelihoods (Insight)
for i, class_label in enumerate(nb.classes_):
    print(f"\nClass {class_label} likelihoods:")
    print(dict(zip(X_train.columns, nb.theta_[i])))



```
